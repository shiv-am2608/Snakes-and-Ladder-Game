```python
import random as rd #to generate random numbers b/w 1-6
snakes={17:7,54:34,62:19,98:79} #given in problem statement{start:end}
ladder={3:38,24:33,42:93,72:84} #given in problem statement{start:end}
```

```python
#to check if the given input is string or number
def check(s):
    return s.isdecimal()
```

```python
print("##### Welcome to Snakes and Ladders Game #####")    #to display welcoming message
player1=input("Enter Name of Player1:")                    #to accept name of Player1
player2=input("Enter Name of Player2:")                    #to accept name of player2
print("###### Let Us Start ######")                        #to mark the start of game
flag,pos1,pos2=0,0,0       #flag->check alternate turns | pos1->position of player1 | pos2->position of player2
while True:                                      #infinite loop untill game ends
    if flag==0:                                  #check for turn of Player1
        print("-----------------------------------------------------") 
        turn=input("Player1:")                   #input from player1
        if(not check(turn)):                     #function calling
            turn=turn.lower()                    #to handle all cases of "Roll" and "Quit" i.e;"roll","ROLL","Roll"...
            if turn=="roll":                     #to check if input is "roll"
                turn=rd.randint(1,6)             #to generate random numbers b/w 1-6
            elif turn=="quit":                   #to check if Player1 wants to "Quit"
                print("Player2 won the game")    #displaying winner if Player1 decided to quit
                break                            #since Player1 quits,to end the game
            else:                                #to display error message for invalid input format
                print("Invalid!!Please enter a number b/w 1-20 or enter roll to genrate a random number b/w 1-6")
                continue                         #to give back the turn to Player1
        else:                                    #to handle the case when input is a number
            turn=int(turn)                       #explicitly converting string to integer
            if(turn==0 or turn>20):              #to check if the given number lies outside the given constraints
                print("Invalid!!Please enter a number b/w 1-20 or enter roll to genrate a random number b/w 1-6")
                continue                         #to give back the turn to Player1
        flag=1                                   #by now,we are confirmed it is a valid move so next turn would be of Player2                          
        print("You got a",turn)                  #to display his valid move
        if pos1+turn>100:                        #to check that Player1 does not overshoot 100
            print("Your Final Position is",pos1) #to display the current position
            continue                             #giving next turn to Player2
        pos1+=turn                               #not overshooting,added to result
        if pos1 in snakes.keys():                #to check if there is a snake
            pos1=snakes[pos1]                    #to update position if there is a snake
        elif pos1 in ladder.keys():              #to check if there is a ladder
            pos1=ladder[pos1]                    #to update position if there is a ladder
        print("Your Final Position is",pos1)     #to display the final position
        if pos1==100:                            #to check for winning condition
            print("Player1 won the Game")        #displaying victory message
            break                                #game completed
    else:                                        #*****Player2 Turn*****
        print("-----------------------------------------------------")
        turn=input("Player2:")                   #input from player2
        if(not check(turn)):                     #function calling
            turn=turn.lower()                    #to handle all cases of "Roll" and "Quit" i.e;"roll","ROLL","Roll"...
            if(turn=="roll"):                    #to check if input is "roll"
                turn=rd.randint(1,6)             #to generate random numbers b/w 1-6
            elif turn=="quit":                   #to check if Player2 wants to "Quit"
                print("Player1 won the game")    #displaying winner if Player2 decided to quit
                break                            #since Player2 quits,to end the game
            else:                                #to display error message for invalid input format
                print("Invalid!!Please enter a number b/w 1-20 or enter roll to genrate a random number b/w 1-6")
                continue                         #to give back the turn to Player2
        else:                                    #to handle the case when input is a number
            turn=int(turn)                       #explicitly converting string to integer
            if(turn==0 or turn>20):              #to check if the given number lies outside the given constraints
                print("Invalid!!Please enter a number b/w 1-20 or enter roll to genrate a random number b/w 1-6")
                continue                         #to give back the turn to Player2
        flag=0                     #by now,we are confirmed it is a valid move so next turn would be of Player2 
        print("You got a",turn)                  #to display his valid move
        if pos2+turn>100:                        #to check that Player2 does not overshoot 100
            print("Your Final Position is",pos2) #to display the current position
            continue                             #giving next turn to Player1
        pos2+=turn                               #not overshooting,added to result
        if pos2 in snakes.keys():                #to check if there is a snake
            pos2=snakes[pos2]                    #to update position if there is a snake
        elif pos2 in ladder.keys():              #to check if there is a ladder
            pos2=ladder[pos2]                    #to update position if there is a ladder
        print("Your Final Position is",pos2)     #to display the final position 
        if pos2==100:                            #to check for winning condition
            print("Player2 won the Game")        #displaying victory message
            break                                #game completed
print("##### Game Successfully Finished #####")  #to mark the end of game
```
